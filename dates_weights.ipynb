```python
import pandas as pd
import numpy as np
import datetime
import calendar
```

```python
c=calendar.TextCalendar(calendar.MONDAY)
```

```python
year=input("Enter the year for the schedule ")
month= input("Enter the month for the schedule ")
```

```python
if(month.lower()=='january' or month.lower()=='Jan' or month==1):
    month_no=1
elif(month.lower()=='february' or month.lower()=='Feb' or month==2):
    month_no=2
elif(month.lower()=='march' or month==3):
    month_no=3
elif(month.lower()=='april'or month==4):
    month_no=4
elif(month.lower()=='may' or month==5):
    month_no=5
elif(month.lower()=='june' or month==6):
    month_no=6
elif(month.lower()=='july' or month==7):
    month_no=7
elif(month.lower()=='august'or month.lower()=='aug' or month==8):
    month_no=8
elif(month.lower()=='september'or month.lower()=='sept' or month==9):
    month_no=9
elif(month.lower()=='october'or month.lower()=='oct' or month==10):
    month_no=10
elif(month.lower()=='november'or month.lower()=='nov' or month==11):
    month_no=11
elif(month.lower()=='december'or month.lower()=='dec' or month==12):
    month_no=12
    
month_no
```

```python
str=c.formatmonth(int(year),int(month_no))
str
```

```python
cal_dates=[]
for i in c.itermonthdays(int(year),int(month_no)):
    cal_dates.append(i)
print(len(cal_dates)) 
cal_dates
```

```python
list_days=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
```

```python
list_month=list_days[0:len(cal_dates)]
len(list_month)
```

```python
weights_days=[]
for i in list_month:
    if(i=='Monday'):
        check=1
        weights_days.append(check)
    elif(i=='Tuesday'):
        check=2
        weights_days.append(check)
    elif(i=='Wednesday'):
        check=1
        weights_days.append(check)
    elif(i=='Thursday'):
        check=2
        weights_days.append(check)
    elif(i=='Friday'):
        check=4
        weights_days.append(check)
    elif(i=='Saturday'):
        check=5
        weights_days.append(check)
    elif(i=='Sunday'):
        check=3
        weights_days.append(check)


print(len(weights_days))        
weights_days
```

```python
dates_result= pd.DataFrame({'Date' : cal_dates, 'Day' : list_month, 'Weight' : weights_days})
dates_result
```
